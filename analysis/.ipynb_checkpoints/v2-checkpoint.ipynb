```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.io as pio
pio.renderers.default = 'iframe'
sns.set(rc={'figure.figsize':(16, 2)})
```

# Характеристики функц-я разработанного модуля на 4000 тиках

```python
df = pd.read_csv("the_only_world_monitor_data_RIGHT_DOWN_4000_at_2025_09_22_11_56_08.csv")
df
```

## Линейность перемещения игрока в бенчмарке

```python
fig = px.line_3d(df, x="player_x", y="player_y", z="tick")
fig.update_layout(
    scene=dict(
        zaxis=dict(range=[0, df.iloc[-1]["tick"]])
    )
)
fig.show()
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(df[['player_x', 'player_y', 'tick']].corr(),
            annot=True, vmax=.3, center=0, square=True,
            linewidths=.5, cbar_kws={"shrink": .5},
            cmap = sns.diverging_palette(230, 20, as_cmap=True),
)
plt.show()
```

## Перевод оси тиков в ось пройденной дистанции

```python
np.sqrt(df.iloc[-1]["player_x"] ** 2 + df.iloc[-1]["player_y"] ** 2)
```

```python
distance_coef = np.sqrt(df.iloc[-1]["player_x"] ** 2 + df.iloc[-1]["player_y"] ** 2) / df.iloc[-1]["tick"]
distance_coef
```

```python
df["distance_pixels"] = df["tick"] * distance_coef
df
```

## Задержка кадра из-за операций чтения/записи с диском

```python
fig,ax = plt.subplots(figsize=(16,2))
sns.lineplot(x="distance_pixels", y="read_write_time_ms", ax=ax, data=df)
ax.set_yticks(np.arange(0, 51, 5))
```

```python
df["read_write_time_ms"].describe()
```

```python
df["read_write_time_ms"].median()
```

## Использование ОЗУ

```python
fig,ax = plt.subplots(figsize=(16,2))
sns.lineplot(x="distance_pixels", y="mem_usage_mb", ax=ax, data=df)
ax.set_yticks(np.arange(55.0, 57.1, 0.25))
```

```python
df["mem_usage_mb"].describe()
```

```python
df["mem_usage_mb"].median()
```

## Динамика чтения/записи фрагментов мира 

```python
plt.figure(figsize=(16, 2))
ax = None
ax = sns.lineplot(x=df['distance_pixels'], y=df['chunks_loaded'], label='chunks_from_disk', ax=ax)
ax = sns.lineplot(x=df['distance_pixels'], y=-df['chunks_unloaded'], label='chunks_to_disk', ax=ax)
ax.set_yticks(np.arange(-7, 18, 7))
plt.show()
```

## Количество тайлов и размер области тайлов в игровом движке 

```python
fig,ax = plt.subplots(figsize=(16,2))
sns.lineplot(x="distance_pixels", y="used_cells", ax=ax, data=df)
ax.set(ylim=(4092, 4100))
```

```python
fig = px.line_3d(df, x="used_cells_bbox_w", y="used_cells_bbox_h", z="distance_pixels")
fig.update_layout(
    scene=dict(
        zaxis=dict(range=[0, df.iloc[-1]["distance_pixels"]])
    )
)
fig.show()
```

## Использование ПЗУ в байтах

```python
fig,ax = plt.subplots(figsize=(16,2))
sns.lineplot(x="distance_pixels", y="savefile_size", ax=ax, data=df)
ax.set_yticks(np.arange(0, 9e6, 1e6))
ax.set_yticks(np.arange(0, 9e6, 1e6))
ax.set_xticks(np.arange(0, 1.8e6, 2.5e5))
```

## Линейная аппроксимация использования ПЗУ

```python
coefs_k_b = np.polyfit(df["distance_pixels"],df["savefile_size"],1)
print(f"k={coefs_k_b[0]}\nb={coefs_k_b[1]}")
fig,ax = plt.subplots(figsize=(16,2))
sns.lineplot(x="distance_pixels", y="savefile_size", ax=ax, data=df, label='savefile_size')
sns.lineplot(x="x", y="y", ax=ax, data=pd.DataFrame({
    "x": [0, 1.75e6],
    "y": [coefs_k_b[1], coefs_k_b[0] * 1.75e6 + coefs_k_b[1]]
}), label='linear_aprox')
ax.set_yticks(np.arange(0, 9e6, 1e6))
ax.set_xticks(np.arange(0, 1.8e6, 2.5e5))
```
